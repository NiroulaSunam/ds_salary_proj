## Sales Analysis

Import Necessary Libraries

```python
import pandas as pd
import os 
import matplotlib.pyplot as plt
```

## Task #1: Merge the 12 months of sales data into a single CSV file

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")

files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
                     
all_months_data.to_csv("all_data.csv", index=False)
    
```

Read in updated datafraom

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

### Cleaning NaN or empty values

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
all_data = all_data.dropna(how='all')
all_data.head()
```

### Find 'Or' and delete it

```python
tem_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
tem_df.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

## Augment Data with additonal colums

### Task 2 : Add Month Column

```python
all_data['month'] = all_data['Order Date'].str[0:2]
all_data['month'] = all_data['month'].astype('int32')
all_data.head()
```

## Task 3: Add a sales column

### Convert columns to correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # Make Int (can also use .astype('int32'))
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # Make Float
```

### Task 3


```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Task 4 : Add a City Column

```python
# Let's use .apply()

"""
def get_city(address):
    return address.split(',')[1]
"""

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1] + ' (' + get_state(x) + ')') # f string i.e f"{x.split(',')[1]} {get_state(x)}
all_data.head()
```

### Question 1: What whas the best month for sales? How much was earned that month?

```python
results = all_data.groupby('month').sum()
results
```

```python
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

### Question 2 : What city had the highest number of sales?

```python
results_city = all_data.groupby('City').sum()
results_city
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results_city['Sales'])
plt.xticks(cities, rotation='vertical', size = 8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City name')
plt.show()
```

### Question 4: What products are most often sold together?

```python
all_data
```

```python
duplicate_data = all_data[all_data['Order ID'].duplicated(keep=False)]
duplicate_data
```

```python
duplicate_data['Grouped'] = duplicate_data.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
duplicate_data = duplicate_data[['Order ID', 'Grouped']].drop_duplicates()
duplicate_data
```

```python
#Counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in duplicate_data['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key,value)
```

### Question 5: What product sold the most? Why do you think it sold the most?

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')

quantity_ordered = product_group.sum()['Quantity Ordered']

quantity_ordered
```

```python
products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products, rotation='vertical', size =8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

print(prices)
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Products Name')
ax1.set_ylabel('Quantity Ordered', color ='g')
ax2.set_ylabel('Price)$)', color = 'b')
ax1.set_xticklabels(products, rotation='vertical', size =8)

plt.show()
```
