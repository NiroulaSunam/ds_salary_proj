# Sales Analysis

**Import Necessary Libraries**

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

### Merge all the sales data

```python
path = "./Sales_Data"
files = [a for a in os.listdir(path) if not a.startswith('.')] #Ignore hidden files
files
```

```python
all_months_data = pd.DataFrame() #create a data Frame to capture the data of all months 
all_months_data
```

```python
for a in files:            #use for loop to read the csv file and save it in current data and concat or add it to the all_months_data and continue it
    current_data = pd.read_csv(path+"/"+a)
    all_months_data = pd.concat([all_months_data, current_data])
```

```python
all_months_data
```

```python
all_months_data.to_csv("all_data.csv", index=False) #Convert and save the dataFrame to csv file 
```

### Read the Combined Data

```python
data = pd.read_csv("all_data.csv")
data
```

### Cleaning Data

Cleaning data here to make it easier to get the question answers at the end of the questions. Leaving some spaces here to find out what problems crop when asking questions and trying to solve it. 

```python
data.dtypes 
#check the numbers and see if it int, float or string ; it can be see that all are string or obj, so convert all nums later when doing calculations
```

**NaN Errors**

```python
data = data.dropna(how='all')
```

```python
data
```

**invalid literal for int() with base 10 Error** - Error where non integer values are in 'Quantity Ordered'

```python
data = data[data['Quantity Ordered'] != 'Quantity Ordered']
data = data[data['Price Each'] != 'Price Each']
```

**Add Months Column**

```python
data['Month'] = data['Order Date'].str[0:2]
data['Month'] = data['Month'].astype('int32')
data
```

**Convert Orderd Received and Price Each to Integer to get the value at the monthly table becaue they are still string**

```python
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'])
data['Price Each'] = pd.to_numeric(data['Price Each'])
```

## Add City Column from Purchase Address for Question 2

```python
def get_city(address):
    return address.split(",")[1]

def get_state(address):
    return address.split(",")[2].split(" ")[1]
```

```python
data['City'] = data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
data.head()
```

## Question 1: What was the best month for sales? How much was earned that month?

```python
data['Sales'] = data['Quantity Ordered'].astype('int') * data['Price Each'].astype('float')
```

```python
data
```

```python
data.groupby(['Month']).sum()
```

```python
months = range(1,13)
print(months)

plt.bar(months,data.groupby(['Month']).sum()['Sales'])
plt.plot(months,data.groupby(['Month']).sum()['Sales'], color='red')
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

```python
values = data.groupby(['Month']).sum()['Sales']
label = range(1,13)

plt.pie(values, labels = values)

plt.title('Monthly Sales')

plt.legend(label,title='Month',bbox_to_anchor=(1.18, 1), loc='upper left', borderaxespad=0)

plt.show()
```

From the bar grah and Pie chart, It can be see that the maximum number of sales was done in the month of December with the Total Sales of 4613443.34 dollars. And the minimum amount of sales was in the year of September with Total Sales of 2097560.13 dollars.

## Question 3 : What time should we display advertisements to maximize likelihood of customer's buying product?

```python
data.head()
```

```python
# Add hour column
data['Hour'] = pd.to_datetime(data['Order Date']).dt.hour
data['Minute'] = pd.to_datetime(data['Order Date']).dt.minute
data['Count'] = 1
data.head()
```

## What products are most often sold together?

```python
# https://stackoverflow.com/questions/43348194/pandas-select-rows-if-id-appear-several-time
df = data[data['Order ID'].duplicated(keep=False)]

# Referenced: https://stackoverflow.com/questions/27298178/concatenate-strings-from-several-rows-using-pandas-groupby
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

```python
product_group = data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = [pair for pair, df in product_group]
plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
# Referenced: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib

prices = data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
